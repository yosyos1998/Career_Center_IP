```python
import pandas as pd
import numpy as np
import tensorflow_hub as hub
import tensorflow_text
import numpy as np
```

```python
# Define the paths to your preprocessed datasets
resume_data_file = '/data/notebook_files/CvDatasetFinal (3).csv'
job_offers_data_file = '/data/notebook_files/JobOffersDatasetFinal1.csv'
```

```python
# Load preprocessed datasets from CSV files using pandas
resumes_df = pd.read_csv(resume_data_file)
job_offers_df = pd.read_csv(job_offers_data_file)
resumes_df.head()
```

```python
# Load pre-trained Universal Sentence Encoder model from TensorFlow Hub
model_url = "https://tfhub.dev/google/universal-sentence-encoder-multilingual/3"
model = hub.load(model_url)
```

```python
sample_df = job_offers_df.sample(frac=0.025, random_state=1)
```

```python
sample_df['job_id'] = range(1, len(sample_df)+1)
resumes_df['resume_id'] = range(1, len(resumes_df)+1)
```

```python
# Encode resume and job offer text using the Universal Sentence Encoder
resume_embeddings = model(resumes_df['Skills'].tolist())
job_embeddings = model(sample_df['Job description'].tolist())
```

```python
import numpy as np
# Compute cosine similarity between resume and job offer embeddings
similarities = np.dot(job_embeddings, np.transpose(resume_embeddings))
```

```python
# Find the best matching resumes for each job offer
top_k = 5 # Number of top matching resumes to recommend
recommendations = {}
for i in range(len(sample_df)):
    job_id = sample_df.iloc[i]['job_id']
    similar_resumes_indices = (-similarities[i, :]).argsort()[:top_k]
    similar_resumes_scores = similarities[i, similar_resumes_indices]
    similar_resumes_ids = resumes_df.iloc[similar_resumes_indices]['resume_id'].tolist()
    recommendations[job_id] = list(zip(similar_resumes_ids, similar_resumes_scores))
```

```python
# Print the recommended resumes for each job offer
for job_id, recommended_resumes in recommendations.items():
    print("Job Offer ID:", job_id)
    for resume_id, score in recommended_resumes:
        print("\tRecommended Resume ID:", resume_id, "with score", score)
```
